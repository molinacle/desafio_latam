```python
# Carga los datos de cada pestaña del archivo US_Regional_Sales_Data.xlsx en un DataFrame independiente.

import pandas as pd

df_sales_orders = pd.read_excel('US_Regional_Sales_Data.xlsx', sheet_name='Sales Orders Sheet')
df_customers = pd.read_excel('US_Regional_Sales_Data.xlsx', sheet_name='Customers Sheet')
df_store_locations = pd.read_excel('US_Regional_Sales_Data.xlsx', sheet_name='Store Locations Sheet')
df_products = pd.read_excel('US_Regional_Sales_Data.xlsx', sheet_name='Products Sheet')
df_sales_team = pd.read_excel('US_Regional_Sales_Data.xlsx', sheet_name='Sales Team Sheet')
```

```python
# Cruza todos los DataFrames usando validación many_to_one, y guarda el resultado en una nueva variable llamada df_base.

df_base = pd.merge(df_sales_orders, df_customers, on='_CustomerID', how='left', validate='many_to_one')
df_base = pd.merge(df_base, df_store_locations, on='_StoreID', how='left', validate='many_to_one')
df_base = pd.merge(df_base, df_products, on='_ProductID', how='left', validate='many_to_one')
df_base = pd.merge(df_base, df_sales_team, on='_SalesTeamID', how='left', validate='many_to_one')
```

```python
df_base.head()
```

```python
# Crea las columnas ProcurementDays, ShippingDays, DeliveryDays y CustomerDays.

df_base['ProcurementDays'] = df_base['OrderDate'] - df_base['ProcuredDate']
df_base['ShippingDays'] = df_base['ShipDate'] - df_base['OrderDate']
df_base['DeliveryDays'] = df_base['DeliveryDate'] - df_base['ShipDate']
df_base['CustomerDays'] = df_base['ShippingDays'] + df_base['DeliveryDays']

df_base.head()
```

```python
# Crea la columna CustomerDaysInterval en base al valor que tenga la columna CustomerDays.

intervalos = [0, 15, 30, 45, 60, 75, 90]
etiquetas = ['0 to 15 days','15 to 30 days','30 to 45 days','45 to 60 days','60 to 75 days','75 to 90 days']

df_base['CustomerDays'] = df_base['CustomerDays'].dt.days.astype('int64')
df_base['CustomerDaysInterval'] = pd.cut(x=df_base['CustomerDays'], bins=intervalos, labels=etiquetas)
```

```python
df_base.head()
```

```python
# Crea el reporte1: Cantidad de ventas por vendedor e Intervalo de Dias al Cliente.

conteo_ordenes = pd.pivot_table(df_base,
                                values='OrderNumber',
                                index='Sales Team',
                                columns='CustomerDaysInterval',
                                aggfunc='count',
                                margins=True,
                                margins_name='Total Ordenes')

print(conteo_ordenes)
```

```python
# Crea la columna de Margen Bruto.

df_base['GrossMargin'] = df_base['Order Quantity']*(df_base['Unit Price']*(1-df_base['Discount Applied'])-df_base['Unit Cost'])
```

```python
# Crea una columna para almacenar el porcentaje de comisión de cada venta.

intervalos = [0, 100, 1000, 10000, 100000]
etiquetas = [0.05, 0.1, 0.15, 0.2]

df_base['CommissionsPercentage'] = pd.cut(x=df_base['GrossMargin'], bins=intervalos, labels=etiquetas, include_lowest=True).astype('float64')
```

```python
df_base.head()
```

```python
# Crea una columna para mostrar el monto de la comisión para cada venta.

df_base['CommissionsAmount'] = df_base['GrossMargin'] * df_base['CommissionsPercentage']
```

```python
df_base.head()
```

```python
# Crea una columna para mostrar el Margen Neto de cada venta.

df_base['NetMargin'] = df_base['GrossMargin'] - df_base['CommissionsAmount']
```

```python
df_base.head()
```

```python
# Crea el reporte2, que muestra el total de Margen Bruto, Comisión y Margen Neto para cada vendedor.

reporte2 = pd.pivot_table(df_base,
                                values=['GrossMargin','NetMargin','CommissionsAmount'],
                                index='Sales Team',
                                aggfunc='sum',
                                margins=True,
                                margins_name='Total')

print(reporte2)
```
