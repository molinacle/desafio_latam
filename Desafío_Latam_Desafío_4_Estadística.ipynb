La base de datos world-data-2023.csv proporciona una gran cantidad de información sobre todos los países del mundo, abarcando una amplia gama de indicadores y atributos. Incluye estadísticas demográficas, indicadores económicos, factores ambientales, métricas de atención médica, estadísticas educativas y mucho más. Con la representación de cada país, este conjunto de datos ofrece una perspectiva global completa sobre diversos aspectos de las naciones, lo que permite análisis en profundidad y comparaciones entre países. Las variables consideradas son:
- Country: Nombre del país.
- Density (P/Km2): Densidad de población medida en personas por kilómetro
cuadrado.
- Abbreviation: Abreviatura o código que representa el país.
- Agricultural Land (%): Porcentaje del área de tierra utilizada para fines agrícolas.
- Land Area (Km2): Área total de tierra del país en kilómetros cuadrados.
- Armed Forces Size: Tamaño de las fuerzas armadas en el país.
- Birth Rate: Número de nacimientos por 1,000 habitantes por año.
- Calling Code: Código de llamada internacional para el país.
- Capital/Major City: Nombre de la capital o ciudad principal.
- CO2 Emissions: Emisiones de dióxido de carbono en toneladas.
- CPI: Índice de Precios al Consumidor, una medida de la inflación y el poder
adquisitivo.
- CPI Change (%): Cambio porcentual en el Índice de Precios al Consumidor en
comparación con el año anterior.
- Currency_Code: Código de moneda utilizado en el país.
- Fertility Rate: Número promedio de hijos nacidos de una mujer durante su vida.
- Forested Area (%): Porcentaje del área de tierra cubierta por bosques.
- Gasoline_Price: Precio de la gasolina por litro en moneda local.
- GDP: Producto Interno Bruto, el valor total de bienes y servicios producidos en el
país.
- Gross Primary Education Enrollment (%): Tasa de inscripción bruta en educación primaria.
- Gross Tertiary Education Enrollment (%): Tasa de inscripción bruta en educación terciaria.
- Infant Mortality: Número de muertes por cada 1,000 nacidos vivos antes de cumplir un año de edad.
- Largest City: Nombre de la ciudad más grande del país.
- Life Expectancy: Número promedio de años que se espera que viva un recién
nacido.
- Maternal Mortality Ratio: Número de muertes maternas por cada 100,000
nacidos vivos.
- Minimum Wage: Nivel de salario mínimo en moneda local.
- Official Language: Idioma(s) oficial(es) hablado(s) en el país.
- Out of Pocket Health Expenditure (%): Porcentaje del gasto total en salud
pagado directamente por los individuos.
- Physicians per Thousand: Número de médicos por cada mil personas.
- Population: Población total del país.
- Population: Labor Force Participation (%): Porcentaje de la población que forma
parte de la fuerza laboral.
- Tax Revenue (%): Ingresos fiscales como porcentaje del PIB.
- Total Tax Rate: Carga tributaria total como porcentaje de las ganancias
comerciales.
- Unemployment Rate: Porcentaje de la fuerza laboral que está desempleada.
- Urban Population: Porcentaje de la población que vive en áreas urbanas.
- Latitude: Coordenada de latitud de la ubicación del país.
- Longitude: Coordenada de longitud de la ubicación del país.

```python
# Importación de librerías y datos.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Pregunta1**
- Carga los datos y genera un dataFrame que excluya las columnas correspondientes a la abreviatura del país, código de llamada, capital, código de moneda, ciudad más grande, lenguaje oficial, latitud y longitud. Considerando estos datos, realiza una
inspección inicial sobre ellos, considerando gráficos. Concluye.

```python
# Cargar los datos
df = pd.read_csv("world-data-2023.csv")

# Excluir las columnas especificadas
columns_to_exclude = ['Abbreviation', 'Calling Code', 'Capital/Major City', 'Currency-Code',
                      'Largest city', 'Official language', 'Latitude', 'Longitude']
df = df.drop(columns=columns_to_exclude)

df.head()
```

```python
# Limpieza de caracteres string
df = df.replace({'\$': '', ',': '', '%': ''}, regex=True)

# Iterar sobre todas las columnas para convertirlas a tipo numérico, excluyendo el nombre del país
columna_a_excluir = 'Country'

for columna in df.columns:
    if columna != columna_a_excluir:
        df[columna] = pd.to_numeric(df[columna])
```

```python
df.describe()
```

```python
# Seleccionar solo las columnas que se desea graficar para la inspección visual
columnas_graficar = ['Birth Rate', 'Fertility Rate', 'Armed Forces size', 'GDP', 'Population', 'Life expectancy', 'Physicians per thousand', 'Infant mortality']
df_graficar = df[columnas_graficar]
```

```python
# Inspección inicial con gráficos
sns.pairplot(df_graficar)
plt.show()
```

```python
# Crear el gráfico solo para las columnas seleccionadas
correlation_matrix = df_graficar.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

**Pregunta 2**
- De las variables codificadas numéricamente, ¿hay pares de ellas en la que parezca
haber correlación? ¿Para cuál(es) de ellas se observa más claramente?

**Respuesta**
- El mapa de calor anterior nos muestra las correlaciones entre todas las variables numéricas. Se observa una correlación positiva entre la Tasa de Fertilidad (Fertility Rate) y la Tasa de Natalidad (Birth Rate), y también una correlación negativa entre la Tasa de Mortalidad infantil (Infant Mortality) y la Esperanza de Vida (Life expectancy).

**Pregunta 3**
- Considera los dos pares de variables anteriores y elimina en cada caso los faltantes y atípicos. Justifica en cada caso tu procedimiento.

```python
# Eliminar valores atípicos y faltantes para Fertility Rate y Birth Rate
df = df.dropna(subset=['Fertility Rate', 'Birth Rate'])
df = df[(df['Fertility Rate'] < df['Fertility Rate'].quantile(0.95)) & (df['Birth Rate'] < df['Birth Rate'].quantile(0.95))]

# Eliminar valores atípicos y faltantes para Infant Mortality y Life Expectancy
df = df.dropna(subset=['Infant mortality', 'Life expectancy'])
df = df[(df['Infant mortality'] < df['Infant mortality'].quantile(0.95)) & (df['Life expectancy'] < df['Life expectancy'].quantile(0.95))]
```

**Pregunta 4**
- Construye un gráfico para verificar la correlación entre estas variables, ahora con los datos preparados. Compara con lo obtenido anteriormente.

```python
# Crear una figura con dos subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Gráfico 1: GDP vs Armed Forces Size
axes[0].scatter(df['Fertility Rate'], df['Birth Rate'])
axes[0].set_xlabel('Fertility Rate')
axes[0].set_ylabel('Birth Rate')
axes[0].set_title('Correlación entre la tasa de fertilidad y la tasa de natalidad')

# Gráfico 2: Physicians per Thousand vs Life Expectancy
axes[1].scatter(df['Infant mortality'], df['Life expectancy'])
axes[1].set_xlabel('Infant mortality')
axes[1].set_ylabel('Life Expectancy')
axes[1].set_title('Correlación entre Mortalidad Infantil y Esperanza de Vida')

# Ajustar el espaciado entre subplots para evitar superposiciones
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

**Pregunta 5**

```python
# Grafico de Dispersión para medir la correlación entre la cantidad de médicos y la esperanza de vida.
plt.scatter(df['Physicians per thousand'], df['Life expectancy'])
plt.xlabel('Médicos por cada 1000 habitantes')
plt.ylabel('Esperanza de Vida')
plt.title('Correlación entre la cantidad de Médicos y la Esperanza de Vida')
plt.show()
```

De hecho en el mapa de calor anterior, se mide la correlación en 0,7. Esto hace sentido dado que a medida que aumenta la cantidad de médicos para la población mayor debiese ser también la esperanza de vida de la población.

**Pregunta 6**
- Construye un gráfico para representar la población total de los países y su producto interno bruto. Prepara los datos y escoge el gráfico adecuado (si es necesario, limpia y/o transforma los datos). Justifica.

```python
# Eliminar valores atípicos y faltantes para Population y GDP
df = df.dropna(subset=['Population', 'GDP'])
df = df[(df['Population'] < df['Population'].quantile(0.95)) & (df['GDP'] < df['GDP'].quantile(0.95))]

plt.scatter(df['Population'], df['GDP'])
plt.xlabel('Population')
plt.ylabel('GDP')
plt.title('Población Total vs Producto Interno Bruto (GDP)')
plt.show()
```

**Conclusión**
- Creo que en este caso no existe necesariamente una correlación entre el total de la población de un país y su producto interno bruto. Es decir, hay países con menos población y que tienen mayor producto interno bruto que otros que tienen más población y viceversa.
